```python
pst.stem('affection')
```

```python
pst.stem('running')  
```

```python
pst.stem('maximum')
```

```python
pst.stem('dropped')
```

```python
words_to_stem = ['give', 'giving','given','gave']

for words in words_to_stem:
    print(words+' : ' + pst.stem(words)) 
```

```python
words_to_stem = ['give', 'giving','given','gave','gaved','thining', 'loving', 'maximum','AR Ansari']

for words in words_to_stem:
    print(words+' : ' + pst.stem(words)) 
```

```python
from nltk.stem import LancasterStemmer
lst = LancasterStemmer()

for words in words_to_stem:
        print(words+' : ' + lst.stem(words)) 
```

```python
from nltk.stem import SnowballStemmer
sbst = SnowballStemmer("english")

for words in words_to_stem:
    print(words+' : ' + sbst.stem(words)) 
```

```python
stemmer = SnowballStemmer('german') 
stemmer.stem('Autobanen')
```

# NER ( Named Entity Recognition)

```python
from nltk import ne_chunk
```

# Embedding ---  Transformer every text will convrt Number


```python
from sklearn.feature_extraction.text import CountVectorizer

sentence = "Data Science and ai Genai has great career Ahead "
vectorizer = CountVectorizer()
vector = vectorizer.fit_transform([sentence])
vector.toarray()

```
